```python
# create important columns

df['AgeGroup'] = pd.cut(df['Age'],
                       bins=[18,30,40,50,60],
                       labels=['18-30','30-40','40-50','50-60'])
```

```python
# Tenture group
df['TenureGroup'] = pd.cut(df['YearsAtCompany'],
                          bins=[0,2,5,10,40],
                          labels=['0-2','3-5','6-10','10+'])
```

```python
# quick validation
df[['Age','AgeGroup','YearsAtCompany','TenureGroup']].head()
```

```python
# Total employees
total = len(df)

# Employees left (Attrition=1)
left = (df['Attrition']==1).sum()

# Employees stayed(Attrition=0)
stayed=(df['Attrition']==0).sum()

# Attrition Percentage
attrition_percentage = (left/total)*100

print("Total employees:", total)
print("Employees stayed:", stayed)
print("Employees left:", left)
print("Attrition %:", round(attrition_percentage, 2))
```

**Overall Attrition**

~16% employees left the company

```python
# Attrition by department
df.groupby('Department')['Attrition'].mean().sort_values(ascending=False) * 100
```

**Department Insight**

Sales has highest attrition (~20.6%)

```python
# Attrition by job role
df.groupby('JobRole')['Attrition'].mean().sort_values(ascending=False) * 100
```

**Age Insight**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
```

```python
# Tenture Analysis

df['TenureGroup'] = pd.cut(df['YearsAtCompany'],
                          bins=[0,2,5,10,40],
                          labels=['0-2','3-5','6-10','10+'])

sns.barplot(x='TenureGroup', y='Attrition', data=df)
plt.title("Attrition by Tenure")
plt.show()
```

## Insight ##
"Employees with 0–2 years of tenure show the highest attrition, indicating early-stage disengagement or onboarding challenges.”

```python
# Business Travel Analysis

plt.figure(figsize=(8,5))
sns.barplot(x='BusinessTravel', y='Attrition', data=df)
plt.title("Attrition by Business Travel")
plt.show()
```

## Insight ##
"Employees who travel frequently have the highest attrition, indicating that frequent travel may contribute to stress, fatigue, and poor work-life balance."

```python
# work life balance analysis

plt.figure(figsize=(8,5))
sns.barplot(x='WorkLifeBalance', y='Attrition', data=df)
plt.title("Attrition by Work-Life Balance")
plt.show()
```

## Insight ##
“Attrition is highest among employees with poor work-life balance, frequent travel, and early tenure (0–2 years). This suggests that workload pressure and lifestyle imbalance are key drivers of attrition.”

```python
# Distance from home analysis

plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df)
plt.title("Distance vs Attrition")
plt.show()
```

## Insight ##
“Employees who left the organization tend to live slightly farther from the workplace compared to those who stayed, suggesting that commuting distance may contribute to attrition.”

```python
# Income Vs Attrition

plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Income vs Attrition")
plt.show()
```

## Insight ##
“Employees who left the organization tend to have lower median monthly income compared to those who stayed, indicating that compensation is an important factor influencing attrition.”

## Conclusion
Attrition is mainly driven by early tenure, workload, and compensation factors.
